#### Друзья делят яблоки между собой. Напишите программу, которая вычислит:


*   сколько целых яблок достанется каждому другу?
*   сколько яблок останется не разделенными?
*   по сколько получит каждый из друзей, если они начнут делить яблоки на части?

На вход подаются два числа: количество друзей и яблок.

На выходе должно быть три строки, каждая должна содержать ответ на поставленный вопрос.

`входные данные:`

`10` 

`8`

`выходные данные:`

`0`

`8`

`0.8`

```python
apples = int(input('сколько у вас яблок: '))
friends = int(input('сколько вас всего друзей: '))
delejka = apples // friends
ostatok = apples % friends
delenie_na_chasti = ostatok / friends
print(f'по {delejka} целых яблок достанется каждому другу')
print(f'{ostatok} яблок останется не разделенными')
print(f'{delenie_na_chasti} получит каждый из друзей, если они начнут делить яблоки на части')
```

#### Длина гипотенузы по заданным катетам треугольника. (Можно почитать [здесь](https://ru.wikipedia.org/wiki/%D0%93%D0%B8%D0%BF%D0%BE%D1%82%D0%B5%D0%BD%D1%83%D0%B7%D0%B0)).

$c^2 = a^2 + b^2$

$c = \sqrt{a^2 + b^2}$


`входные данные:`

`3` 

`4`

`выходные данные:`

`25.0`

```python
katet_a = int(input('введите длину катета a: '))
katet_b = int(input('введите длину катета b: '))
print('гипотенузa треугольника с заданными катетами равна: ', katet_a*katet_a+katet_b*katet_b)
```

#### Дано трехзначное число. Найдите сумму его цифр.

`Входные данные`

Вводится целое положительное число.

`Выходные данные`

Выведите ответ на задачу.

**Примеры**

`входные данные`

`179`

`выходные данные`

`17`

`подсказка: 1+7+9 = 17`

```python
a = int(input('enter a number: '))
split_a = str(a)
b = []

for i in split_a:
    b.append(int(i))
print(sum(b))
```
